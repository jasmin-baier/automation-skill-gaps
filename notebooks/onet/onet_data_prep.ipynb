```python
import pandas as pd
```

```python
tasks_to_dwa_path = '/Users/alejandracastaneda/Dropbox/Automation and Skill Survey/03_Data/01_Raw Data/01_ONET ESCO ISCO crosswalk/ONET data/Tasks to DWAs.xlsx'
dwa_reference_path = 'data/DWA Reference.xlsx'
```

```python
tasks_to_dwa = pd.read_excel(tasks_to_dwa_path, engine='openpyxl')
```

```python
dwa_reference = pd.read_excel(dwa_reference_path)
```

```python
tasks_to_dwa
```

```python
dwa_reference
```

```python
joined_df = tasks_to_dwa.merge(dwa_reference, on='DWA ID', how='outer')
```

```python
joined_df
```

```python
# remove the _y columns and rename the _x columns to their original names
joined_df = joined_df[[col for col in joined_df.columns if not col.endswith('_y')]]
joined_df.columns = [col[:-2] if col.endswith('_x') else col for col in joined_df.columns]

```

```python
joined_df
```

```python
# get the number of unique DWAs
num_unique_dwas = joined_df['DWA ID'].nunique()
print(f'Number of unique DWAs: {num_unique_dwas}')
```

```python
# get number of duplicates based on DWA ID
num_duplicates = joined_df.duplicated(subset=['DWA ID']).sum()
print(f'Number of duplicate DWA IDs: {num_duplicates}')
```

```python
# get number of duplicate tasks
task_duplicates = joined_df.duplicated(subset=['Task ID']).sum()
print(f"Number of duplicate Task IDs: {task_duplicates}")
```

```python
# get number of duplicates for all columns
total_duplicates = joined_df.duplicated().sum()
print(f'Number of total duplicate rows: {total_duplicates}')
```

```python
# find if there is any column with a unique id
for col in joined_df.columns:
    if joined_df[col].is_unique:
        print(f'Column {col} has unique values.')
```

```python
joined_df.nunique()
```

```python
dwa_task_duplicates = joined_df.duplicated(subset=['DWA ID', 'Task ID']).sum()
print(f'Number of duplicate DWA ID and Task ID combinations: {dwa_task_duplicates}')
```
